### Write a Pandas program to sort the &#39;cut&#39; column in ascending order and return it in a Series

```python
sorted_df = df.sort_values(by='cut', ascending=True)
ser = pd.Series(sorted_df.cut)
print(ser)
```

### Write a Pandas program to find the diamonds that are with a Fair or Good or Premium.

```python
sd = sorted_df[sorted_df['cut'].isin(['Fair', 'Good', 'Premium'])]
print(sd)
```
